–ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø—Ä–æ—Ö–æ–¥–∏—Ç —Ç–∞–∫–æ–π –ø—É—Ç—å –∫ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—é –¥–æ—Å—Ç–∞–≤–∫–∏ –±–ª—é–¥–∞ –∏–∑ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤:
1. `authorization` ‚Äî –∞–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è ‚Äî –∑–∞–ø—É—Å–∫ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è.
2. `main_page` ‚Äî –∑–∞–≥—Ä—É–∑–∫–∞ –æ—Å–Ω–æ–≤–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è.
2. `reg_page` ‚Äî –ø–µ—Ä–µ—Ö–æ–¥ –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü—É —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏.
2. `confirm_phone` ‚Äî –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ —Ç–µ–ª–µ—Ñ–æ–Ω–∞ (–æ–∫–æ–Ω—á–∞–Ω–∏–µ –ø—Ä–æ—Ü–µ—Å—Å–∞ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏).
2. `object_page` ‚Äî —Å—Ç—Ä–∞–Ω–∏—Ü–∞ –±–ª—é–¥–∞.
2. `add_to_cart` ‚Äî –ø–µ—Ä–µ—Ö–æ–¥ –≤ –∫–æ—Ä–∑–∏–Ω—É.
2. `order` ‚Äî –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ –∑–∞–∫–∞–∑–∞.    

````python
# === –®–ê–ì 1: –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –ù–û–í–´–• –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ===
df_new = df[df['first_date'].dt.date == df['date']].copy()

# === –®–ê–ì 2: –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —ç—Ç–∞–ø–æ–≤ –≤–æ—Ä–æ–Ω–∫–∏ ===
funnel_steps = [
    'authorization',
    'main_page',
    'reg_page',
    'confirm_phone',
    'object_page',
    'add_to_cart',
    'order'
]

# –§–∏–ª—å—Ç—Ä—É–µ–º —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—ã–µ —Å–æ–±—ã—Ç–∏—è
df_funnel = df_new[df_new['event'].isin(funnel_steps)].copy()

# === –®–ê–ì 3: –ü–æ–¥—Å—á—ë—Ç —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ ===
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ —Å–æ–±—ã—Ç–∏—é –∏ —Å—á–∏—Ç–∞–µ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö visitor_uuid
funnel_counts = df_funnel.groupby('event')['visitor_uuid'].nunique().reindex(funnel_steps, fill_value=0)

# –°–æ–∑–¥–∞—ë–º DataFrame
funnel_df = pd.DataFrame({
    'step': funnel_steps,
    'users': funnel_counts.values
})

# –£–¥–∞–ª–∏–º —ç—Ç–∞–ø—ã —Å –Ω—É–ª—ë–º, –µ—Å–ª–∏ –≤–¥—Ä—É–≥ –æ–Ω–∏ –µ—Å—Ç—å (–¥–ª—è —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏)
funnel_df = funnel_df[funnel_df['users'] > 0].reset_index(drop=True)

# === –®–ê–ì 4: –†–∞—Å—á—ë—Ç –∫–æ–Ω–≤–µ—Ä—Å–∏–π ===
# –ö–ª–∞—Å—Å–∏—á–µ—Å–∫–∞—è –≤–æ—Ä–æ–Ω–∫–∞: % –æ—Ç –ø–µ—Ä–≤–æ–≥–æ —à–∞–≥–∞
funnel_df['conversion_from_first_%'] = (funnel_df['users'] / funnel_df['users'].iloc[0] * 100).round(1)

# Step-by-step: % –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —à–∞–≥–∞
funnel_df['conversion_from_previous_%'] = (funnel_df['users'] / funnel_df['users'].shift(1) * 100).round(1)
````

````python
# –°–æ–∑–¥–∞—ë–º –æ—Ç–¥–µ–ª—å–Ω—ã–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –¥–ª—è –º–æ–±–∏–ª—å–Ω—ã—Ö –∏ –¥–µ—Å–∫—Ç–æ–ø–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
df_mobile = df_funnel[df_funnel['device_type'] == 'Mobile']
df_desktop = df_funnel[df_funnel['device_type'] == 'Desktop']
````

````python
# –î–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—á–∏—Ç–∞–µ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –∫–∞–∂–¥–æ–º —ç—Ç–∞–ø–µ ===
def build_funnels_by_device(df_device):
    funnel_counts = {}
    for step in funnel_steps:
        count = df_device[df_device['event'] == step]['visitor_uuid'].nunique()
        funnel_counts[step] = count
    return pd.DataFrame({
        'step': funnel_steps,
        'users': [funnel_counts[step] for step in funnel_steps]
    })

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≤–æ—Ä–æ–Ω–æ–∫
mobile_funnels = build_funnels_by_device(df_mobile)
desktop_funnels = build_funnels_by_device(df_desktop)

# –†–∞—Å—á—ë—Ç –∫–æ–Ω–≤–µ—Ä—Å–∏–π step-by-step 
def calculate_step_conversions(funnel_df):
    funnel_df['conversion_from_previous_%'] = (funnel_df['users'] / funnel_df['users'].shift(1) * 100).round(1)
    return funnel_df

mobile_funnels = calculate_step_conversions(mobile_funnels)
desktop_funnels = calculate_step_conversions(desktop_funnels)
````

````python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è ‚Äî —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ step-by-step –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–ª—è –º–æ–±–∏–ª—å–Ω–æ–π –∏ –¥–µ—Å–∫—Ç–æ–ø–Ω–æ–π –≤–µ—Ä—Å–∏–π ===

# –§–∏–ª—å—Ç—Ä—É–µ–º, —á—Ç–æ–±—ã —É–±—Ä–∞—Ç—å –ø–µ—Ä–≤—ã–π —à–∞–≥ (–Ω–µ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ)
mobile_plot = mobile_funnels.iloc[1:].copy()
desktop_plot = desktop_funnels.iloc[1:].copy()

x = np.arange(len(mobile_plot['step']))  # –ø–æ–∑–∏—Ü–∏–∏ —Å—Ç–æ–ª–±—Ü–æ–≤
width = 0.35  # —à–∏—Ä–∏–Ω–∞ —Å—Ç–æ–ª–±—Ü–æ–≤

fig, ax = plt.subplots(figsize=(12, 6))

# –°—Ç–æ–ª–±—Ü—ã –¥–ª—è –º–æ–±–∏–ª—å–Ω–æ–π –≤–µ—Ä—Å–∏–∏
rects1 = ax.bar(x - width/2, mobile_plot['conversion_from_previous_%'], width, label='Mobile', color='lightcoral')

# –°—Ç–æ–ª–±—Ü—ã –¥–ª—è –¥–µ—Å–∫—Ç–æ–ø–Ω–æ–π –≤–µ—Ä—Å–∏–∏
rects2 = ax.bar(x + width/2, desktop_plot['conversion_from_previous_%'], width, label='Desktop', color='lightblue')

# –ü–æ–¥–ø–∏—Å–∏
ax.set_title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ step-by-step –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –º–µ–∂–¥—É –º–æ–±–∏–ª—å–Ω–æ–π –∏ –¥–µ—Å–∫—Ç–æ–ø–Ω–æ–π –≤–µ—Ä—Å–∏—è–º–∏')
ax.set_xlabel('–≠—Ç–∞–ø—ã –≤–æ—Ä–æ–Ω–∫–∏')
ax.set_ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —à–∞–≥–∞ (%)')
ax.set_xticks(x)
ax.set_xticklabels(mobile_plot['step'], rotation=45, ha='right')
ax.legend()

# –ü–æ–¥–ø–∏—Å—ã–≤–∞–µ–º —Å—Ç–æ–ª–±–∏–∫–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}%', xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
````

````python
# –í—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã —Å –¥–∞–Ω–Ω—ã–º–∏ ===
print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã step-by-step –∫–æ–Ω–≤–µ—Ä—Å–∏–∏:")
comparison_table = pd.DataFrame({
    'step': mobile_plot['step'],
    'mobile_conversion_%': mobile_plot['conversion_from_previous_%'],
    'desktop_conversion_%': desktop_plot['conversion_from_previous_%']
})
print(comparison_table)
````

````python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è RFM-–∞–Ω–∞–ª–∏–∑–∞:
df_rfm = df_filtered.groupby('user_id').agg(
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π —Å –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–∫–∞–∑–∞:    
    recency = ('order_recency', lambda x: x.min().days),
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –∑–∞ –ø–µ—Ä–∏–æ–¥ –≤—Ä–µ–º–µ–Ω–∏: 
    frequency = ('datetime', 'count'),
    # –°—É–º–º–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –≤—Å–µ—Ö –∑–∞–∫–∞–∑–æ–≤: 
    monetary_value = ('revenue', 'sum')
).reset_index()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
print(df_rfm.head(3))
````

````python
# –°—Ç—Ä–æ–∏–º —Ç—Ä–∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
fig, axes = plt.subplots(3, 1, figsize=(15, 15))

# –î–∞–≤–Ω–æ—Å—Ç—å
axes[0].hist(df_rfm['recency'], bins=75)
axes[0].set_title('Recency (–¥–Ω–∏ —Å –ø–æ—Å–ª–µ–¥–Ω–µ–π –ø–æ–∫—É–ø–∫–∏)')
axes[0].set_xlabel('–î–Ω–∏')
axes[0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')

# –ß–∞—Å—Ç–æ—Ç–∞
axes[1].hist(df_rfm['frequency'], bins=75)
axes[1].set_title('Frequency (—á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤)')
axes[1].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
axes[1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')

# –°—É–º–º–∞ –≤—Å–µ—Ö –∑–∞–∫–∞–∑–æ–≤
axes[2].hist(df_rfm['monetary_value'], bins=75)
axes[2].set_title('Monetary (—Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–æ–≤)')
axes[2].set_xlabel('–°—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–æ–≤')
axes[2].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')

plt.show()
````

**–ê–Ω–∞–ª–∏–∑ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º**

**1. Recency (–¥–Ω–∏ —Å –ø–æ—Å–ª–µ–¥–Ω–µ–π –ø–æ–∫—É–ø–∫–∏)**
- **–§–æ—Ä–º–∞:** –ü–∏–∫ –Ω–∞ –Ω—É–ª–µ–≤—ã—Ö –¥–Ω—è—Ö (–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∫—É–ø–∞–ª–∏ `03.07.2021`), –∑–∞—Ç–µ–º —Å–ø–∞–¥.
- **–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ:** `~60 –¥–Ω–µ–π` ‚Üí –∑–Ω–∞—á–∏—Ç, –µ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø–æ–∫—É–ø–∞–ª–∏ –±–æ–ª–µ–µ `2 –º–µ—Å—è—Ü–µ–≤`.
- **–í—ã–≤–æ–¥:** –ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∞–∫—Ç–∏–≤–Ω—ã –≤ –ø–æ—Å–ª–µ–¥–Ω–∏–µ `30‚Äì45` –¥–Ω–µ–π, –Ω–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–∞—è —á–∞—Å—Ç—å (–æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ—Å–ª–µ `45` –¥–Ω–µ–π) ‚Äî `"—Å–ø—è—â–∏–µ"` –∏–ª–∏ `"–ø–æ—Ç–µ—Ä—è–Ω–Ω—ã–µ"`.

**2. Frequency (—á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤)**
- **–§–æ—Ä–º–∞:** –†–µ–∑–∫–∏–π –ø–∏–∫ –Ω–∞ `1` –∑–∞–∫–∞–∑–µ, –∑–∞—Ç–µ–º —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–π —Å–ø–∞–¥.
- **–ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (–±–æ–ª–µ–µ `4500` –∏–∑ `~6000`)** —Å–æ–≤–µ—Ä—à–∏–ª–∏ —Ç–æ–ª—å–∫–æ `1` –∑–∞–∫–∞–∑.
- **–õ–∏—à—å –Ω–µ–±–æ–ª—å—à–∞—è —á–∞—Å—Ç—å** ‚Äî `2`, `3`, `5+` –∑–∞–∫–∞–∑–æ–≤.
- **–í—ã–≤–æ–¥:** –°–µ—Ä–≤–∏—Å –∏–º–µ–µ—Ç –æ—á–µ–Ω—å –≤—ã—Å–æ–∫—É—é –¥–æ–ª—é –æ–¥–Ω–æ—Ä–∞–∑–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π ‚Äî —ç—Ç–æ –≥–ª–∞–≤–Ω–∞—è –ø—Ä–æ–±–ª–µ–º–∞

**3. Monetary (—Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –∑–∞–∫–∞–∑–æ–≤)**
- **–§–æ—Ä–º–∞:** –ê–Ω–∞–ª–æ–≥–∏—á–Ω–∞ Frequency ‚Äî —Ä–µ–∑–∫–∏–π –ø–∏–∫ –Ω–∞ –Ω–∏–∑–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö (–¥–æ 5000 —Ä—É–±.), –∑–∞—Ç–µ–º —Å–ø–∞–¥.
- **–ú–Ω–æ–≥–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø–æ—Ç—Ä–∞—Ç–∏–ª–∏ –º–∞–ª–æ (–¥–æ `1000‚Äì5000 —Ä—É–±.`)**, **–µ–¥–∏–Ω–∏—Ü—ã** ‚Äî –±–æ–ª—å—à–µ `20 000`.
- **–í—ã–≤–æ–¥:** –ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Äî `–Ω–∏–∑–∫–æ–±—é–¥–∂–µ—Ç–Ω—ã–µ`, –∏ –¥–∞–∂–µ —Ç–µ, –∫—Ç–æ —Å–¥–µ–ª–∞–ª –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–∞–∫–∞–∑–æ–≤, –Ω–µ —Ç—Ä–∞—Ç—è—Ç –º–Ω–æ–≥–æ.

**–ó–∞–¥–∞—á–∞ 3.2.** –í—ã–¥–µ–ª–∏—Ç–µ **RFM-—Å–µ–≥–º–µ–Ω—Ç—ã** –∏ –∏–∑—É—á–∏—Ç–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –Ω–∏–º.
    
–ü—Ä–∏–º–µ–Ω–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –ø–æ–¥—Ö–æ–¥—ã –∫ –≤—ã–¥–µ–ª–µ–Ω–∏—é RFM —Å–µ–≥–º–µ–Ω—Ç–æ–≤:
- **Recency** ‚Äî —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –Ω–∞ —Ç—Ä–∏ —Ä–∞–≤–Ω—ã–µ –≥—Ä—É–ø–ø—ã.
- **Frequency** ‚Äî  —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –Ω–∞ —Ç—Ä–∏ –≥—Ä—É–ø–ø—ã: 
    - –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 1 –∑–∞–∫–∞–∑–æ–º (F-—Å–µ–≥–º–µ–Ω—Ç —Å –Ω–µ–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤);
    - –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 2 –∑–∞–∫–∞–∑–∞–º–∏ (F-—Å–µ–≥–º–µ–Ω—Ç —Å–æ —Å—Ä–µ–¥–Ω–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤);
    - –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å 3 –∏ –±–æ–ª–µ–µ –∑–∞–∫–∞–∑–∞–º–∏ (F-—Å–µ–≥–º–µ–Ω—Ç —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤).
- **Monetary** ‚Äî —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –Ω–∞ —Ç—Ä–∏ –≥—Ä—É–ø–ø—ã –ø–æ 33-–º—É –∏ 66-–º—É –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º.
    
–ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–º —Å–µ–≥–º–µ–Ω—Ç–µ, —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã–≤–æ–¥—ã –∏ –Ω–∞–ø–∏—à–∏—Ç–µ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏.

````python
# –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –ø–æ Recency ‚Äî –†–ê–í–ù–´–ï –ò–ù–¢–ï–†–í–ê–õ–´ (pd.cut)
min_r = df_rfm['recency'].min()
max_r = df_rfm['recency'].max()

# –°–æ–∑–¥–∞—ë–º 3 —Ä–∞–≤–Ω—ã—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: [min, min+step), [min+step, min+2*step), [min+2*step, max]
step = (max_r - min_r) / 3
bins = [min_r - 1e-5, min_r + step, min_r + 2 * step, max_r + 1e-5]  # –Ω–µ–±–æ–ª—å—à–æ–π –∑–∞–ø–∞—Å –¥–ª—è –≤–∫–ª—é—á–µ–Ω–∏—è –≥—Ä–∞–Ω–∏—Ü

df_rfm['r'] = pd.cut(
    df_rfm['recency'],
    bins=bins,
    labels=[3, 2, 1],  # 3 ‚Äî —Å–∞–º—ã–µ —Å–≤–µ–∂–∏–µ, 1 ‚Äî —Å–∞–º—ã–µ –¥–∞–≤–Ω–∏–µ
    include_lowest=True
)
````

````python
# –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –ø–æ Frequency ‚Äî –ø–æ –ø—Ä–∞–≤–∏–ª–∞–º
df_rfm['f'] = df_rfm['frequency'].apply(
    lambda x: 1 if x == 1 else (2 if x == 2 else 3)
)
````

````python
# –°–µ–≥–º–µ–Ω—Ç–∞—Ü–∏—è –ø–æ Monetary ‚Äî –ø–æ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º (q=33%, 66%)
df_rfm['m'] = pd.qcut(
    df_rfm['monetary_value'].rank(method='first'),  # –æ–±—Ö–æ–¥ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
    q=[0, 0.33, 0.66, 1],
    labels=[1, 2, 3]
)
````

````python
# –ù–∞–π—Ö–æ–¥–∏–º –≥—Ä—É–ø–ø–æ–≤–æ–π RFM-–∏–Ω–¥–µ–∫—Å:
df_rfm[['r','f','m']] = df_rfm[['r','f','m']].astype('str')
df_rfm['rfm_group'] = df_rfm['r'] + df_rfm['f'] + df_rfm['m']
````

````python
# –í—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω –ø–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ —Ç–∞–±–ª–∏—Ü—ã:
print(df_rfm.head(3))
````

````python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º –∏ —Å—á–∏—Ç–∞–µ–º –∏—Ö —Ä–∞–∑–º–µ—Ä:
rfm_group = df_rfm.groupby('rfm_group')['user_id'].nunique().reset_index()

# –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –¥–æ–ª—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–º —Å–µ–≥–º–µ–Ω—Ç–µ:
rfm_group['user_share'] = rfm_group['user_id'] / rfm_group['user_id'].sum()

# –í—ã–≤–µ–¥–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç:
print(rfm_group.sort_values(by='user_id', ascending=False))
````

### –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π –≤—ã–≤–æ–¥

- –¢–æ–ø-3 —Å–µ–≥–º–µ–Ω—Ç–∞ ‚Äî `211`, `111`, `311` ‚Äî –≤—Å–µ –∏–º–µ—é—Ç `F = 1`, —Ç–æ –µ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å–æ–≤–µ—Ä—à–∏–ª–∏ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –∑–∞–∫–∞–∑.

- –ù–µ —Ç—Ä–∞—Ç–∏—Ç—å –±—é–¥–∂–µ—Ç –Ω–∞ –≤–æ–∑–≤—Ä–∞—Ç, –∞ –∏–Ω–≤–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ —É–¥–µ—Ä–∂–∞–Ω–∏–µ –∏ —Ä–æ—Å—Ç LTV:

  - –í–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –≤–ª–∏—è—Ç—å –Ω–∞ –º–µ–Ω—é (–æ–ø—Ä–æ—Å—ã, —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ).

## 4. K-Means –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è

–ß—Ç–æ–±—ã –ª—É—á—à–µ –ø–æ–Ω–∏–º–∞—Ç—å –ø–æ–≤–µ–¥–µ–Ω–∏–µ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –∏ –ø—Ä–µ–¥–ª–∞–≥–∞—Ç—å –∏–º –ø–µ—Ä—Å–æ–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —É—Å–ª–æ–≤–∏—è —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞, –∞ —Ç–∞–∫–∂–µ —Å–æ–∑–¥–∞–≤–∞—Ç—å –ø–µ—Ä—Å–æ–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –º–∞—Ä–∫–µ—Ç–æ–ª–æ–≥–∏ –∑–∞–ø—Ä–æ—Å–∏–ª–∏ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º:
- –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤;
- —Å—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞;
- –æ–±—â–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–∞–Ω–Ω—ã—Ö —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –±–ª—é–¥ (–∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞).

–° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ K-Means —Ä–∞–∑–¥–µ–ª–∏—Ç–µ —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã –Ω–∞ –∫–ª–∞—Å—Ç–µ—Ä—ã, –æ—Ç—Ä–∞–∂–∞—é—â–∏–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –º–æ–¥–µ–ª–∏ —Ä–∞–±–æ—Ç—ã –∏ —Ü–µ–ª–µ–≤—ã–µ —Å–µ–≥–º–µ–Ω—Ç—ã. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–¥–µ–ª–∏—Ç—å –≥—Ä—É–ø–ø—ã —Ä–µ—Å—Ç–æ—Ä–∞–Ω–æ–≤ —Å –ø–æ—Ö–æ–∂–∏–º–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞–º–∏ –∏ —Ä–∞–∑—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –¥–ª—è –Ω–∏—Ö —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–µ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã–µ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏.

**–ó–∞–¥–∞—á–∞ 4.1.** –ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞ –ø–æ–¥—Å—á–∏—Ç–∞–π—Ç–µ:
- –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤;
- —Å—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞;
- –æ–±—â–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–∞–Ω–Ω—ã—Ö —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –±–ª—é–¥ (–∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞).

````python
# –§–∏–ª—å—Ç—Ä—É–µ–º —Ç–æ–ª—å–∫–æ –∑–∞–∫–∞–∑—ã (–≥–¥–µ –µ—Å—Ç—å order_id)
orders = df[df['event'] == 'order'].copy()

# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ rest_id –∏ object_id ‚Äî —Å—Ç—Ä–æ–∫–∏ (–∏–ª–∏ —Ö–æ—Ç—è –±—ã –Ω–µ NaN —Ç–∞–º, –≥–¥–µ –Ω—É–∂–Ω–æ)
orders = orders.dropna(subset=['rest_id'])
````


<div class="alert alert-danger">
    
  <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</b>

  ‚ùå  K-Means –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –ø—Ä–æ–≤–æ–¥—è—Ç —Ç–∞–∫-–∂–µ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–∫–∞–∑–æ–≤, —Ç.–µ. –Ω–∞–º –Ω–∞–¥–æ –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ, –æ—Å—Ç–∞–≤–∏–≤ —Ç–æ–ª—å–∫–æ —Å–æ–±—ã—Ç–∏–µ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ –∑–∞–∫–∞–∑–∞ - 'order'. 
    
–û—Ç—Ñ–∏–ª—å—Ç—Ä—É–π –¥–∞–Ω–Ω—ã–µ –∏ –ø—Ä–æ–≤–µ—Ä—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—Å–µ–≥–æ —à–∞–≥–∞ K-Means –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏.
    
</div>

<div class="alert alert-block alert-info">
  
  <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π c—Ç—É–¥–µ–Ω—Ç–∞</b>
    
 –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∞.

<div class="alert alert-success">
  
  <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v.3</b>
    
  ‚úîÔ∏è –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –≤—ã–±—Ä–∞–ª–∏ –≤–µ—Ä–Ω–æ.
 
</div> 

````python
# –ê–≥—Ä–µ–≥–∞—Ü–∏—è –ø–æ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞–º
restaurant_metrics = orders.groupby('rest_id').agg(
    order_count=('order_id', 'count'),
    avg_order_value=('revenue', 'mean'),
    unique_dishes=('object_id', 'nunique')  # —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –±–ª—é–¥–∞
).reset_index()
````

````python
# –£–¥–∞–ª–∏–º —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã —Å –Ω—É–ª–µ–≤—ã–º–∏/–Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
restaurant_metrics = restaurant_metrics[
    (restaurant_metrics['order_count'] > 0) &
    (restaurant_metrics['avg_order_value'] > 0) &
    (restaurant_metrics['unique_dishes'] > 0)
].reset_index(drop=True)

print(restaurant_metrics.head())
````

**–ó–∞–¥–∞—á–∞ 4.2.** –ù–æ—Ä–º–∞–ª–∏–∑—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ –∏ –Ω–∞–π–¥–∏—Ç–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–≤ –º–µ—Ç–æ–¥ `.inertia_`. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∏–Ω–µ—Ä—Ü–∏–∏ –æ—Ç —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (–º–µ—Ç–æ–¥ ¬´–ª–æ–∫—Ç—è¬ª) –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

````python
# –í—ã–±–∏—Ä–∞–µ–º —Ç–æ–ª—å–∫–æ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏
features = ['order_count', 'avg_order_value', 'unique_dishes']
X = restaurant_metrics[features]
````

````python
# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python
# –ú–µ—Ç–æ–¥ "–ª–æ–∫—Ç—è"
inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o')
plt.title('–ú–µ—Ç–æ–¥ "–ª–æ–∫—Ç—è" –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤')
plt.xlabel('–ß–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
````
